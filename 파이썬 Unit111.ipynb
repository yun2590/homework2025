```
names={'Mary':10999,'Cris':2111,'Aimy':9778,'Tom':20245}
name=input('이름을 입력하세요: ')
if name in names:
  print(f'이름이 <{name}>인 출생아수는 <{names[name]:,d}>명입니다.')
else:
  print(f'딕셔너리 names에 <{name}>인 이름이 존재하지 않습니다.')
```
