```
mystack=[]
```

```
def putData(data):
  mystack.append(data)
```

```
def popData():
  if len(mystack) == 0:
    return None
  return mystack.pop()
```

```
putData('사과')
putData([1,2,3])
putData(3.141592)
print(f'스택상태: {mystack}')
```

```
ret=popData()
while ret != None:
  print(f'스택에서 데이터 추출: {ret}')
  print(f'스택상태: {mystack}')
  print('----------------')
  ret=popData()
```
